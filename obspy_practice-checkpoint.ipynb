## 1.python的介绍
其中glob是通配符模块
obspy中的read，用于读取各种格式的地震数据

```python
import glob
from obspy.core import read
for file in glob.glob('*.SAC'):
    st=read(file)
    tr=st[0]
    msg="%s %s %f %f" %(tr.stats.station,str(tr.stats.starttime),tr.data.mean(),tr.data.std())
    print(msg)
    #print(st[0].data)
```

## 2.UTCDateTime
世界标准时间数据

```python
from obspy.core import UTCDateTime
UTCDateTime("2012-09-07T12:12:00")
t1=UTCDateTime(1997,12,2,20)
t1=t1-8*3600
t2=UTCDateTime()
t3=UTCDateTime(0)
print(t1,t2,(t2-t1)//3600,'\n',t1.julday,t1.timestamp,t1.weekday,t3,t2+86400)
```

```python
#用于创建一个列表，从昨天10：00开始，相隔90分钟
t=[]
for i in range(10):
    t0=UTCDateTime().date
    t1=UTCDateTime(t0)-14*3600
    t2=t1+90*60*i
    #print(type(str(t2)))
    t.append(t2)
print(t)
```

波形绘制
单道分量和三道分量分别读取绘制

```python
from obspy.core import read
single=read('http://examples.obspy.org/COP.BHZ.DK.2009.050')
print(single)

three=read('http://examples.obspy.org/COP.BHZ.DK.2009.050')
three+=read('http://examples.obspy.org/COP.BHN.DK.2009.050')
three+=read('http://examples.obspy.org/COP.BHE.DK.2009.050')
print(three)
```

```python
#画图，以及画图参数的指定
single[0].plot()
three.plot()
three.plot(size=(800,500))
dt=single[0].stats.starttime
single.plot(color='red',number_of_ticks=5,tick_rotation=10,tick_format='%I:%M %p',starttime=dt+60*60,endtime=dt+60*60+120)
single.plot(outfile='singlechannel.png')
```

```python
#使用matplotlib自定义绘图
import matplotlib.pyplot as plt 
from obspy import read 
st = read() 
tr = st[0] 
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1) 
ax.plot(tr.times("matplotlib"), tr.data, "b-") 
ax.xaxis_date() 
fig.autofmt_xdate() 
plt.show()
```

## 滤波

```python
import numpy as np
import matplotlib.pyplot as plt
import obspy
st = read("http://examples.obspy.org/RJOB_061005_072159.ehz.new")
tr=st[0]
tr_filt=tr.copy()
tr_filt.filter('bandpass',freqmin=0.01,freqmax=1,corners=4)
#plot
t=np.arange(0,tr.stats.npts/tr.stats.sampling_rate,tr.stats.delta)

plt.figure(figsize=[10,6])
plt.subplot(211)
plt.plot(t,tr.data,'k')
plt.ylabel('Raw Data')
plt.subplot(212)
plt.plot(t,tr_filt.data,'k')
plt.ylabel('Bandpassed Data')
plt.xlabel('Time[s]')
plt.suptitle(tr.stats.starttime)
fig=plt.gcf()
plt.show(fig)
fig.savefig('bandpass.svg')
```

## 降采样

```python
import obspy 
import numpy as np
import matplotlib.pyplot as plt
st=read('http://examples.obspy.org/RJOB_061005_072159.ehz.new')
tr=st[0]
tr_new=tr.copy()
tr_new.decimate(factor=4)
tr_filt=tr.copy()
tr_filt.filter('lowpass',freq=0.4*tr.stats.sampling_rate/4.0)


#plot
t=np.arange(0,tr.stats.npts/tr.stats.sampling_rate,tr.stats.delta)
t_new=np.arange(0,tr_new.stats.npts/tr_new.stats.sampling_rate,tr_new.stats.delta)
t_filt=np.arange(0,tr_filt.stats.npts/tr_filt.stats.sampling_rate,tr_filt.stats.delta)
plt.figure(dpi=300)
plt.plot(t,tr.data,'k',label='Raw',alpha=0.3)
plt.plot(t_new,tr_new.data,'b',label='Decimate',alpha=0.3)
plt.plot(t_filt,tr_filt.data,'r',label='Lowpass',alpha=0.3)
plt.xlabel('Time [s]')
plt.xlim([82,83.5])
plt.suptitle(tr.stats.starttime)
plt.legend()
plt.savefig('decimate.png')
plt.show()
```
